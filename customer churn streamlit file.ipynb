```python
import pandas as pd
import numpy as np
from datetime import timedelta
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import streamlit as st
```

```python
st.title("Model Deployment: 👤📉Customer churn Prediction")
```

```python
st.sidebar.header("User Input Features")
```

```python
def input_features():
    event_time = st.sidebar.text_input()
    event_type = st.sidebar.text_input()
    price = st.sidebar.number_input()
    user_id = st.sidebar.number_input()
    user_session = st.sidebar.text_input()
    dic = {'event_time': event_time,
    'event_type':event_type,
    'price':price,
    'user_id':user_id,
    'user_session':user_session}
    data = pd.DataFrame(dic,index =[0])
    return data
result = input_features()
```
